```python
# Count the amount of values for Yes and No
counts = data['Response'].value_counts()
counts
```

```python
# Show a plot of the total number of responses.
counts = data['Response'].value_counts().plot(kind='bar')
```

```python
# Show a plot of the response rate by the sales channel.
sns.countplot(x="Sales Channel", hue="Response", data=data)
```

```python
# Show a plot of the response rate by the total claim amount.
sns.countplot(x="Total Claim Amount", hue="Response", data=data)
```

```python
# Show a plot of the response rate by income.
sns.countplot(x = "Response", hue = "Income", data = data)
```
